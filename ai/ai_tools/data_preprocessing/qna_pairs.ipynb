```python
import pandas as pd
```

```python
path = r"C:\All University Materials\Project\ICT304-project\ai\ai_tools\data_preprocessing\DATA_BIG.json"
df = pd.read_json(path, lines=True)

df.columns
```

```python
df = df.drop(columns=['topic', 'question_title'])
df = df.rename(columns={'question_body' : 'Question', 'answer_body' : 'Answer'})
df
```

```python
import html
import re

def clean_for_ai(text):
    
    if not isinstance(text, str):
        return ""

    # 1. Unescape HTML entities (&lt; becomes <, &amp; becomes &, etc.)
    text = html.unescape(text)

    # 2. Normalize newlines
    # Replace 3 or more consecutive newlines with exactly 2 (to maintain paragraph separation)
    text = re.sub(r'\n{3,}', '\n\n', text)

    # 3. Strip leading and trailing whitespace
    text = text.strip()

    return text

df['Question'] = df['Question'].apply(clean_for_ai)
df['Answer'] = df['Answer'].apply(clean_for_ai)
df
```

```python
df.to_json('stack_overflow_qna_sample.json',orient='records', lines=True)
```
